# Pandas Serileri

```python
#liste üzerinden pandas serisi oluşturma
a=pd.Series([1,2,3,4,5])
print(a)
print(type(a))
print("seri özellikleri :",a.axes)
print("serinin tipi",a.dtype)
print("seri boşmu:",a.empty)
print("serinin eleman sayısı:",a.size)
print("serinin içindeki değerler:",a.values)
print("serinin ilk iki elemanı:",a.head(2)) #veya a[0:3] head baştan bakmak için kullanılır
print("serinin son iki elemanı:",a.tail(2)) #tail sondan bakmak için kullanılır
```

```python
#NumPy array üzerinden pandas serisi oluşturma
a=np.array([1,25,32,14,45])

seri=pd.Series(a)
print(seri)
```

```python
#oluşturulan serinin indexlerini belirleme
seri=pd.Series([1,2,3,7,0.1],index=[1,2,3,4,0])
print(seri)
```

```python
#index değerleri sayı olmak zorunda değildir
seri=pd.Series([1,2,3,7,0.1],index=[list("abcde")]) #veya index=["a","b","c","d","e"]
print(seri)
print("a indexindeki eleman:",seri["a"].values)
```

```python
#sözlük üzerinden pandas serisi oluşturma
sozluk={"elma":3,"portakal":5,"muz":3.5}
seri=pd.Series(sozluk)
print(seri)
print("seri elma:>",seri["elma"])
```

## DataFrame Birleştirme İşlemleri

```python
s1=np.random.randint(10,size=5)
s2=np.random.randint(10,size=5)
s3=np.random.randint(10,size=5)
df1=pd.DataFrame({"var1":s1,"var2":s2,"var3":s3})
df2=df1+99
```

```python
df1
```

```python
df2
```

```python
#satır bazında birleştirme işlemi
df3=pd.concat([df1,df2])
df3
```

```python
#sutun bazında birleştirme için
df3=pd.concat([df1,df2],axis=1)
df3
```

```python
#satır bazında birleştirme işlemi yaptığımı zaman index sorunu ortaya çıkıyor
#ignoe_index parametresini kullanarak bu sorunu halledebiliriz
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

```python
#sütun isimleri farklı olan DataFramleri birleştirme
df2.columns=["var1","var2","deg3"]
df3=pd.concat([df1,df2])
df3
```

```python
#ortaya çıkan NaN değerlerini kullanmıyorsak sadece kesişen değerleri kullanmak için farklı bir parametre kullanmamız gerekiyor
df3=pd.concat([df1,df2],join="inner")
df3
```

```python
#birleştirme işlemleri bir dataframi baz alarak yapma
df3=pd.concat([df1,df2],join_axes=[df1.columns])
df3
```

### Aggregatefilter,transform,apply şlemleri

```python
df=pd.DataFrame({"gruplar":["A","B","C","A","B","C"],
                "degisken1":[10,23,56,98,78,20],
                "degisken2":[100,255,655,254,321,789]},
               columns=["gruplar","degisken1","degisken2"])
df
```

```python
df.groupby("gruplar").aggregate(["min","median",max])
```

```python
df.groupby("gruplar").aggregate({"degisken1":"std",
                                "degisken2":"median"})
```

```python
#filter işlemi
def filter_func(x):
    return x["degisken1"].std()>30
```

```python
df.groupby("gruplar").filter(filter_func)
```

```python
#transform fonksiyonu
df.groupby("gruplar").transform(lambda x:x-x.mean())
```

```python
#Apply fonksiyonu
df.groupby("gruplar").apply(np.sum)
```

```python
L=[0,1,0,1,2,0]
df.groupby(L).sum()
```

### Pivot Tablolar

```python
import pandas as pd
import seaborn as sns
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.groupby("sex")[["survived"]].mean()
```

```python
df.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#tek boyutlu pivot
df.pivot_table("survived",index="sex",columns="class")
```

```python
#çok boyutlu pivot ,satırlara boyut ekleme
age=pd.cut(df["age"],[0,18,90])
age.head()
```

```python
df.pivot_table("survived",["sex",age],"class")
```

```python
#sütunlara boyut ekleme
fare=pd.qcut(df["fare"],2)
fare.head()
```

```python
df.pivot_table("survived",["sex",age],[fare,"class"])
```

```python
df.pivot_table(index="sex",columns="class",
              aggfunc={"survived":sum,"fare":"mean"})
```

```python
df.pivot_table("survived",index="sex",columns="class",margins=True)
```
